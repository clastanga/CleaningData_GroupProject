```python
size = pd.DataFrame(data['size'].value_counts())
size = size.reset_index()
size.columns = ['size', 'counts']
size
```

```python
def clean_size(text):
        if text == '10001+':
            return 'very big'
        elif text == '5001-10000':
            return 'big'
        elif text == '1001-5000':
            return 'big'
        elif text == '501-1000':
            return 'medium'
        elif text == '201-500':
            return 'medium'
        else:
            return 'small'

size['size'] = list(map(clean_size, size['size']))
```

```python
size
```

```python
data['size'] = list(map(clean_size, data['size']))
```
